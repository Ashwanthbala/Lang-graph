```python
from typing import Dict, TypedDict, List
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):
    values: List[int]
    name: str
    operation: str
    fresult: str

def processing_node(state: AgentState) -> AgentState:
    """  Function that processes value and prints the result """
    if state["operation"] == "+":
        state["fresult"] = f"Hi {state["name"]}, Your answer is {sum(state["values"])}"
    if state["operation"] == "*":
        ans = 1
        for res in state["values"]:
            ans *= res
        state["fresult"] = f"Hi {state["name"]}, Your answer is {ans}"
    return state
            
    
```

```python
graph = StateGraph(AgentState)
graph.add_node("processor", processing_node)
graph.set_entry_point("processor")
graph.set_finish_point("processor")

app = graph.compile()
```

```python
result = app.invoke({"name": "Ashwanth","operation": "*","values": [1,2,3,4]})
```
